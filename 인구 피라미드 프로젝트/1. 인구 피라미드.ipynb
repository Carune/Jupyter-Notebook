# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
file_name = '202207_202207_연령별인구현황_월간.xlsx'
file_name2 = '201207_201207_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, E:Y')
df_m2 = pd.read_excel(file_name2, skiprows=3, index_col='행정기관', usecols='B, E:Y')
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int) # 콤마를 제거하고 정수형으로 변환
df_m2.iloc[0] = df_m2.iloc[0].str.replace(',', '').astype(int)
```

## 여자 데이터 정의

```python
df_w = pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
df_w2 = pd.read_excel(file_name2, skiprows=3, index_col='행정기관', usecols='B, AB:AV')
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w2.columns = df_m2.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int) # 콤마를 제거하고 정수형으로 변환
df_w2.iloc[0] = df_w2.iloc[0].str.replace(',', '').astype(int) 
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False 
```

```python
plt.figure(figsize = (10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2022 대한민국 인구 피라미드')
plt.savefig('2022_인구 피라미드.png', dpi=100)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 15))

axs[0].barh(df_m2.columns, -df_m2.iloc[0] // 1000)
axs[0].barh(df_w2.columns, df_w2.iloc[0] // 1000)
axs[0].set_title('2012 대한민국 인구 피라미드')
axs[0].set(xlabel = '인구 수 (단위 : 천 명)', ylabel = '나이')

axs[1].barh(df_m.columns, -df_m.iloc[0] // 1000)
axs[1].barh(df_w.columns, df_w.iloc[0] // 1000)
axs[1].set_title('2022 대한민국 인구 피라미드')
axs[1].set(xlabel = '인구 수 (단위 : 천 명)', ylabel = '나이')
```
